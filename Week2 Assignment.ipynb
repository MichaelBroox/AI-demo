### Question 1. Creating a table using Pandas

```python
# importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# creating a Dataframe
df = pd.DataFrame ( {'Name':['John', 'Mary', 'Peter', 'Jeff', 'Bill', 'Lisa', 'Jose'],
                    'Age':[23, 78, 22, 19, 45, 33, 20],
                    'Gender':['Male', 'Female', 'Male', 'Male', 'Male', 'Female', 'Male'],
                    'Region':['Bono', 'Bono East', 'Bono', 'Bono East', 'Bono', 'Ahafo', 'Ahafo'],
                    'Number_of_Children':[2, 0, 0, 3, 2, 1, 4],
                    'Number_of_Pets':[5, 1, 0, 5, 2, 2, 3]} )
```

### Question 2a.

```python
# plotting a scatter graph
plt.figure(figsize=(10, 6))
plt.scatter(df.Number_of_Children, df.Number_of_Pets)
plt.title('Graph of Number of Children and Number of Pets', fontsize = 22) # graph title
plt.xlabel('Number of Children', fontsize = 12) # x-axis label
plt.ylabel('Number of Pets', fontsize = 12) # y-axis label
plt.show()
```

### Question 2b.

```python
# plotting a bar graph
plt.figure(figsize=(10, 6))
plt.bar(df.Name, df.Age, color = 'green')
plt.title('Graph of Name and Age', fontsize = 22) # graph title
plt.xlabel('Name', fontsize = 12) # x-axis label
plt.ylabel('Age', fontsize = 12) # y-axis label
plt.show()
```

### Question 2c.

```python
# plotting short-dashes line graph
plt.figure(figsize=(10, 6))
plt.plot(df.Name, df.Number_of_Pets, 'ro--', lw=5) # plotting the line graph with short red dashes
plt.title('Graph of Name and Number of Pets', fontsize = 22) # graph title
plt.xlabel('Name', fontsize = 12) # x-axis label
plt.ylabel('Number of Pets', fontsize = 12) # y-axis label
plt.show()
```

```python
import urllib3 as url
from bs4 import BeautifulSoup
import requests as re
```

```python
http = url.PoolManager()
```

```python
website = http.request('GET', 'https://en.wikipedia.org/wiki/Web_scraping') # getting data from the website
web_content = BeautifulSoup(website.data, 'lxml') # using beautifulsoup to make website data readable

print(web_content.title) # printing the title of the web content
print(web_content.title.text) # printing the title of the web content as readable text
print(web_content.text) # printing the whole web content as readable text
```
