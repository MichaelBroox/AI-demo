# PLOTTING HARD CODED DATA

```python
n = 5

men = (20, 35, 30, 35, 27)
women = (25, 32, 34, 20, 25)

ind = np.arange(n)
width = 0.35

plt.bar(ind, men, width, Label='Men')
plt.bar(ind, men, width, Label='Women')

plt.ylabel('Scores')
plt.title('Scores by gender')

plt.xticks(ind + width / 2, ('G1','G2','G3', 'G4', 'G5'))

plt.legend(loc='best')
plt.show()
```
